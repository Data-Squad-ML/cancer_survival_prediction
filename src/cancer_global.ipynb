```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/global_cancer_patients_2015_2024.csv')
```

```python
display(data.head())
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
sns.histplot(data['Age'], bins=30, kde=True, ax=axs[0])
axs[0].set_title('Age Distribution')
data['Country_Region'].value_counts().plot.pie(ax=axs[1], autopct='%1.1f%%')
data['Year'].value_counts().plot.pie(ax=axs[2], autopct='%1.1f%%')
plt.tight_layout()
plt.ylabel('')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
sns.histplot(data['Air_Pollution'], bins=30, kde=True, ax=axs[0, 0])
axs[0, 0].set_title('Air Pollution Distribution')
sns.histplot(data['Alcohol_Use'], bins=30, kde=True, ax=axs[0, 1])
axs[0, 1].set_title('Alcohol Use Distribution')
sns.histplot(data['Smoking'], bins=30, kde=True, ax=axs[1, 0])
axs[1, 0].set_title('Smoking Distribution')
sns.histplot(data['Obesity_Level'], bins=30, kde=True, ax=axs[1, 1])
axs[1, 1].set_title('Obesity Level Distribution')
plt.tight_layout()
plt.show()
```

```python
# visualization for Cancer_Type	Cancer_Stage	Treatment_Cost_USD
fig, axs = plt.subplots(1, 3, figsize=(12, 6))
data['Cancer_Type'].value_counts().plot.pie(ax=axs[0], autopct='%1.1f%%')
axs[0].set_title('Cancer Type Distribution')
sns.histplot(data['Treatment_Cost_USD'], bins=30, kde=True, ax=axs[1])
axs[1].set_title('Treatment Cost Distribution')
data['Cancer_Stage'].value_counts().plot.pie(autopct='%1.1f%%', ax=axs[2])
axs[2].set_title('Cancer Stage Distribution')
plt.tight_layout()
plt.ylabel('')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(data['Survival_Years'], bins=30, kde=True, ax=axs[0])
axs[0].set_title('Survival Years Distribution')
sns.histplot(data['Target_Severity_Score'], bins=30, kde=True, ax=axs[1])
axs[1].set_title('Target Severity Score Distribution')
plt.tight_layout()
plt.show()
```

```python
# groupby target_severity_score intervals show distribution of survival years
plt.figure(figsize=(10, 6))
data['Severity_Bin'] = pd.cut(data['Target_Severity_Score'], bins=[0, 2, 4, 6, 8, 10], labels=['0-2', '3-4', '5-6', '7-8', '9-10'])
sns.boxplot(x='Severity_Bin', y='Survival_Years', data=data)
plt.title('Survival Years by Target Severity Score Bins')
plt.xlabel('Target Severity Score Bins')
plt.ylabel('Survival Years')
plt.show()
```
